```python
from pathlib import Path
import matplotlib.pyplot as plt
import plotly.express as px
import pandas as pd
import csv
```

```python
path = Path('weather_data/sitka_weather_07-2021_simple.csv')
lines = path.read_text().splitlines()

reader = csv.reader(lines)
header_row = next(reader)
# print(header_row)

# for index, colum_header in enumerate(header_row):
#   print(index, colum_header)

dates = []
highs = []
lows = []
for row in reader:
  date = row[2]
  high = int(row[4])
  low = int(row[5])
  dates.append(date)
  highs.append(high)
  lows.append(low)
print(highs)

plt.style.use('seaborn-v0_8-dark')
fig, ax = plt.subplots()
ax.plot(highs, c='red')
ax.plot(lows, c='blue')
ax.set_title('Daily Temperature', fontsize=24)
ax.set_ylabel('Temperature', fontsize=16)
ax.set_xlabel('')
# ax.tick_params(labelsize=16)
ax.set_aspect('equal')
ax.grid(True)
plt.show()
```

```python
path = Path('weather_data/sitka_weather_07-2021_simple.csv')
lines = path.read_text().splitlines()

reader = csv.reader(lines)
header_row = next(reader)

# for index, colum_header in enumerate(header_row):
#   print(index, colum_header)

highs = []
lows = []
dates = []
for row in reader:
  date = row[2]
  high = int(row[4])
  low = int(row[5])
  dates.append(date)
  highs.append(high)
  lows.append(low)
# print(dates)
# print(highs)
# print(lows)

data = pd.DataFrame({'Date': dates, 'Highs': highs, 'Lows': lows})
data_long = pd.melt(
  data,
  id_vars=['Date'],
  value_vars=['Highs', 'Lows'],
  var_name='Type',
  value_name='Temperature'
)

fig = px.line(data_long, x='Date', y='Temperature', color='Type', title='Highs and Lows Over Time')


# fig.update_layout(xaxis_dtick=1)

fig.show()
```
